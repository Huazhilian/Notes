- ### Exception handling

**try...except**

```python
while True:
    try:
        x=int(input('Enter a number: '))
        print('The number is:', x)
        break
    except ValueError:
        print('Invalid input, please enter a number.')
```

```python
while True: #always run
    try:
        x=int(input('Enter a number: '))
        print('The number is:', x)
        break #exit loop
    except (TypeError, NameError, ValueError): #exception handling
        pass ##ignore error
```

```python
try:
    f = open('myfile.txt')
except Exception as e:
    print(f"An error occurred: {e}")
```

**try...except...else...finally**

```python
def prob(): # This function raises an exception
    x=1/0
try: # This block may raise an exception
    prob()
except ZeroDivisionError as e: # This block handles the exception
    print('error:',e)
else: # This block executes if no exception occurs
    print('No errors occurred')
finally: # This block always executes
    print('This block always executes')
```

- ### Raise exception

**raise**

```python
def delta(x):
    if x == 0:
        raise Exception("value not defined")
    else:
        return 0
print(delta(1)) # This will work
    
```

**self-defined exception**

```python
class MyError(Exception):
        def __init__(self, value):
            self.value = value
        def __str__(self):
            return repr(self.value)
try:
    raise MyError('s')
except MyError as e:
    print('MyError:', e.value)
```

**assert**

```python
import random
x = random.randint(1, 10)
assert(x<5), "x is not less than 5"
print(x)
```
